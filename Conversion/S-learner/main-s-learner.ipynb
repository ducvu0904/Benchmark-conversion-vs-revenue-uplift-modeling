```python
import pandas as pd
import numpy as np
import optuna
from lightgbm import LGBMClassifier
from sklift.models import SoloModel

import sys
sys.path.append("../..")
from metrics import auuc, auqc, lift, krcc

# 1. Tải dữ liệu
print("Loading data...")
%time train_df = pd.read_csv(r"../../dataset/Hillstrom/Men/train_men.csv")
%time test_df =  pd.read_csv(r"../../dataset/Hillstrom/Men/test_men.csv")
%time val_df = pd.read_csv(r"../../dataset/Hillstrom/Men/val_men.csv")

in_features = ['recency', 'history_segment', 'history', 'mens', 'womens',
               'zip_code', 'newbie', 'channel_Multichannel', 'channel_Phone', 'channel_Web']
label_feature = 'conversion'
treatment_feature = 'treatment'

# Chuẩn bị X, y, treatment cho Train
X_train = train_df[in_features]
y_train = train_df[label_feature]
t_train = train_df[treatment_feature]

# Chuẩn bị X, y, treatment cho Validation (Dùng cho Optuna)
X_val = val_df[in_features]
y_val_true = val_df[label_feature].values.flatten()
t_val_true = val_df[treatment_feature].values.flatten()

# Chuẩn bị X, y, treatment cho Test (Dùng đánh giá cuối cùng)
X_test = test_df[in_features]
y_test_true = test_df[label_feature].values.flatten()
t_test_true = test_df[treatment_feature].values.flatten()


# 2. Định nghĩa hàm Objective cho Optuna
def objective(trial):
    # Khai báo không gian tìm kiếm hyperparameter cho LightGBM
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.1, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'num_leaves': trial.suggest_int('num_leaves', 10, 150),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 200),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'random_state': 42,
        'verbose': -1 # Tắt cảnh báo của LightGBM trong quá trình tuning
    }
    
    # Khởi tạo Base Model và S-Learner
    base_model = LGBMClassifier(**params)
    s_learner = SoloModel(estimator=base_model)
    
    # Huấn luyện trên tập Train
    s_learner.fit(X=X_train, y=y_train, treatment=t_train)
    
    # Dự đoán trên tập Validation
    uplift_pred_val = s_learner.predict(X_val)
    
    # Tính toán AUQC trên tập Validation (Lưu ý: để plot=False để không in hình khi tuning)
    score = auqc(y_val_true, t_val_true, uplift_pred_val, bins=100, plot=False)
    
    return score

# 3. Chạy Optuna Optimization
print("🔃🔃🔃 Bắt đầu chạy Optuna Tuning 🔃🔃🔃")
# Chọn direction='maximize' vì chúng ta muốn AUQC càng lớn càng tốt
study = optuna.create_study(direction="maximize", study_name="S_Learner_LGBM_Tuning")
# Chạy 50 trials (bạn có thể tăng lên 100-200 nếu có nhiều thời gian)
study.optimize(objective, n_trials=50, show_progress_bar=True)

print("✅ Tuning hoàn tất!")
print(f"Giá trị AUQC tốt nhất trên tập Val: {study.best_value:.4f}")
print("Bộ tham số tốt nhất:")
for key, value in study.best_params.items():
    print(f"    {key}: {value}")


# 4. Huấn luyện lại mô hình với bộ tham số tốt nhất & Đánh giá trên tập TEST
print("\n🔃🔃🔃 Huấn luyện mô hình Final với Best Params 🔃🔃🔃")
# Thêm random_state và verbose vào bộ best_params
best_params = study.best_params
best_params['random_state'] = 42
best_params['verbose'] = -1

# Khởi tạo lại model với best params
final_base_model = LGBMClassifier(**best_params)
final_s_learner = SoloModel(estimator=final_base_model)

# Train mô hình final
final_s_learner.fit(X=X_train, y=y_train, treatment=t_train)

# Dự đoán trên tập TEST
uplift_pred_test = final_s_learner.predict(X_test)

# Đánh giá các metrics trên tập TEST
print("\n" + "=" * 40)
print("[FINAL S-LEARNER CONVERSION - TEST SET]")
auuc_score = auuc(y_test_true, t_test_true, uplift_pred_test, bins=100, plot=True)
auqc_score = auqc(y_test_true, t_test_true, uplift_pred_test, bins=100, plot=True)
lift_score = lift(y_test_true, t_test_true, uplift_pred_test, h=0.3)
krcc_score = krcc(y_test_true, t_test_true, uplift_pred_test, bins=100)

print(f"AUUC: {auuc_score:.3f}")
print(f"AUQC: {auqc_score:.3f}")
print(f"Lift: {lift_score:.3f}")
print(f"KRCC: {krcc_score:.3f}")
print("=" * 40)
```
